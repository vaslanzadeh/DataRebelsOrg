## explore year coverage for each data

```python
def find_missing_years(df, df_name="DataFrame"):
    # Convert to integers
    if df_name in ["life_expectancy_df", "health_indicators_df"]:
        years = df['Year']
    else:
        years = list(map(int, df.columns[2:]))

    # Find min and max
    min_year, max_year = min(years), max(years)

    # Find missing years
    missing_years = [year for year in range(min_year, max_year + 1) if year not in years]

    print(f"{df_name}: Min year: {min_year}, Max year: {max_year}")
    print(f"{df_name}: Missing years: {missing_years}\n")

# Call function with DataFrames
find_missing_years(life_expectancy_df, 'life_expectancy_df')
find_missing_years(health_indicators_df, 'health_indicators_df')

find_missing_years(freq_use_green_df, 'freq_use_green_df')
find_missing_years(statisfaction_distance_df, 'statisfaction_distance_df')

find_missing_years(statisfaction_distance_df, 'statisfaction_distance_df')
find_missing_years(use_of_facilities_df, 'use_of_facilities_df')
find_missing_years(walking_distance_df, 'walking_distance_df')
```
