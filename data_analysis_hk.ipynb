# Analysing the datasets

```python
# check the Scottish areas (councils) in the life expectancy dataset
life_scottish_areas = life_expectancy_df[(life_expectancy_df['Country'] == 'Scotland') & (life_expectancy_df['Area type'] == 'Local Areas')].sort_values(['Area name'])
print(life_scottish_areas['Area name'].unique())
```

```python
# check the Scottish areas (councils) in the health survey dataset
health_scottish_areas = health_indicators_df[health_indicators_df['FeatureType'] == 'Council Area'].sort_values(['FeatureName'])
print(health_scottish_areas['FeatureName'].unique())
```

```python
# create a new dataframe for the life expectancy in Scottish council areas between 2015-2023
scot_la_life_expectancy_df = life_expectancy_df[(life_expectancy_df['Year'].isin([2015,2016,2017,2018,2019,2020,2021,2022,2023])) & (life_expectancy_df['Country'] == 'Scotland') & (life_expectancy_df['Area type'] == 'Local Areas')]
```

## Male life expectancy in Scottish council areas

```python
# get life expectancy for male newborns in each Scottish council area
scot_la_male_le = scot_la_life_expectancy_df[(scot_la_life_expectancy_df['Sex'] == 'Male') & (scot_la_life_expectancy_df['Age band'] == 1)]
print(scot_la_male_le[['Area name', 'Life expectancy', 'Year']])
```

```python
# create a figure
plt.figure(figsize=(12, 6))
plt.title('Life Expectancy for Male Newborns in Scottish Council Areas')
        
# get council area names
areas = scot_la_male_le['Area name'].unique()
        
for area in areas:
    # filter data for the current area
    area_data = scot_la_male_le[scot_la_male_le['Area name'] == area]
            
    # pivot the data (Year vs Age Band)
    pivot_data = area_data.pivot_table(
        index='Year',
        columns='Age band',
        values='Life expectancy',
        aggfunc='mean'  # Handles duplicates
    )
            
    # combine all age groups by averaging the life expectancy across age bands
    pivot_data['All Ages'] = pivot_data.mean(axis=1)
            
    # plot the combined life expectancy for all age groups for the current area
    plt.plot(pivot_data.index, pivot_data['All Ages'], label=area, marker='o')
        
# add labels and legend for each plot
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.legend(title='Area Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
        
# show the plot for this sex
plt.show()
```

```python
# get the highest life expectancy
male_highest_le = scot_la_male_le[['Area name', 'Life expectancy', 'Year']].sort_values('Life expectancy', ascending=False).head(10)
print(male_highest_le)
```

```python
# get the lowest life expectancy
male_lowest_le = scot_la_male_le[['Area name', 'Life expectancy', 'Year']].sort_values('Life expectancy', ascending=False).tail(10)
print(male_lowest_le)
```

```python
# get the average life expectancy for male newborns in each Scottish council area between 2015-2023, sorting from 
male_avg_le = scot_la_male_le['Life expectancy'].groupby(scot_la_male_le['Area name']).mean()
print(male_avg_le.sort_values(ascending=False))
```
